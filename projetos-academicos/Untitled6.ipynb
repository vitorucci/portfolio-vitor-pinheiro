```python
estoque = {
    "banana": 10,
    "maçã": 8,
    "laranja": 4,
}
print("estoque atual: ")
for produto, quantidade in estoque.items():
  print (f"{produto}: {quantidade}")
print()

produto = "banana"
quantidade = 18
if produto in estoque:
  estoque[produto] = quantidade
  print(f"Quantidade do produto {produto} alterada para {quantidade}")
else:
  print(f"o produto {produto} nao existe no estoque")
print(estoque)
```
