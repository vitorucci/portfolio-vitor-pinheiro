```python
dia_aniversario = int(input("digite o dia do seu aniversario (1-31): "))
mes_aniversario = int(input("digite o mes do seu aniversario (1-12): "))

dia_hoje = int(input("digite o dia de hoje(1-31): "))
mes_hoje = int(input("digite o mes de hoje(1-12): "))

dias_por_mes = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

def calcular_dia_do_ano(dia, mes):
  return sum(dias_por_mes[:mes-1]) + dia

hoje_dia_do_ano = calcular_dia_do_ano(dia_hoje, mes_hoje)
aniv_dia_do_ano = calcular_dia_do_ano(dia_aniversario, mes_aniversario)

if hoje_dia_do_ano == aniv_dia_do_ano:
    print("Feliz aniversário!")
elif hoje_dia_do_ano < aniv_dia_do_ano:
    dias_faltando = aniv_dia_do_ano - hoje_dia_do_ano
    print(f"Faltam {dias_faltando} dia(s) para o seu aniversário.")
else:
    dias_passaram = hoje_dia_do_ano - aniv_dia_do_ano
    print(f"Já se passaram {dias_passaram} dia(s) desde o seu aniversário.")
```

```python
v1 = float(input("Digite o primeiro valor positivo: "))
v2 = float(input("Digite o segundo valor positivo: "))
v3 = float(input("Digite o terceiro valor positivo: "))
v4 = float(input("Digite o quarto valor positivo: "))
media_aritmetica = (v1 + v2 + v3 + v4) / 4

media_geometrica = (v1 * v2 * v3 * v4) ** (1/4)

media_harmonica = 4 / ((1/v1) + (1/v2) + (1/v3) + (1/v4))

print("\n--- RESULTADOS ---")
print(f"Média Aritmética:  {media_aritmetica:.2f}")
print(f"Média Geométrica:  {media_geometrica:.2f}")
print(f"Média Harmônica:   {media_harmonica:.2f}")
```

```python
x1 = float(input("Digite o 1º número: "))
x2 = float(input("Digite o 2º número: "))
x3 = float(input("Digite o 3º número: "))
x4 = float(input("Digite o 4º número: "))

media = (x1 + x2 + x3 + x4) / 4
variancia = ((x1 - media)**2 + (x2 - media)**2 + (x3 - media)**2 + (x4 - media)**2) / 4
desvio_padrao = variancia ** 0.5

print("\n--- RESULTADOS ---")
print(f"Média aritmética:  {media:.2f}")
print(f"Variância:         {variancia:.2f}")
print(f"Desvio padrão:     {desvio_padrao:.2f}")
```

```python
sexo = input("Digite seu sexo (homem/mulher): ").strip().lower()
peso = float(input("Digite seu peso (em kg): "))
altura = float(input("Digite sua altura (em cm): "))
idade = int(input("Digite sua idade (em anos): "))

if sexo == 'homem':
    tmb = 66.47 + (13.75 * peso) + (5.003 * altura) - (6.755 * idade)
elif sexo == 'mulher':
    tmb = 655.09 + (9.563 * peso) + (1.85 * altura) - (4.676 * idade)
else:
    print("Sexo inválido!")
    tmb = 0

if tmb > 0:
    print(f"Sua Taxa Metabólica Basal (TMB) é: {tmb:.2f} calorias por dia.")
```
