```python
explicacoes = {
    "int": "int: É um tipo de dado que representa números inteiros, como 1, 2, 100, -5.",
    "float": "float: É um tipo de dado que representa números decimais, como 3.14, -0.5, 2.0.",
    "str": "str: É um tipo de dado que representa textos (strings), como 'olá', '123', 'Python'.",
    "bool": "bool: É um tipo de dado que representa valores lógicos: True (verdadeiro) ou False (falso).",
    "list": "list: É uma coleção de itens, como [1, 2, 3] ou ['a', 'b', 'c']."
}

tipo = input("Qual tipo de variável você deseja aprender? ").strip().lower()

if tipo in explicacoes:
    print(explicacoes[tipo])
else:
    print("Tipo de variável não reconhecido. Tente: int, float, str, bool ou list.")
```

```python
lista1 = ['maçã', 'banana', 'laranja', 'uva']
lista2 = ['melancia', 'abacaxi']

print("Lista 1:", lista1)
print("Lista 2:", lista2)

elemento = input("Qual elemento você deseja mover da Lista 1 para a Lista 2? ")

if elemento in lista1:
    lista1.remove(elemento)
    lista2.append(elemento)
    print(f"Elemento '{elemento}' movido com sucesso!")
else:
    print(f"O elemento '{elemento}' não está na Lista 1.")

print("Conteúdo da Lista 2 após a mudança:", lista2)
```

```python
import random

def jogo_adivinhacao():
    numero_aleatorio = random.randint(1, 10)

    tentativas = 0
    acerto = False

    while not acerto:
        palpite = int(input("Digite um número de 1 a 10: "))

        tentativas += 1


        if palpite == numero_aleatorio:
            acerto = True
            print(f"Você acertou!\nNúmero de tentativas: {tentativas}")
        else:
            print("Tente novamente!")

jogo_adivinhacao()
```
