2)	 Faça um programa que peça números inteiros ao usuário e conta quantos números positivos e quantos números negativos foram digitados. O programa deverá parar quando o número zero for digitado, mostrando na tela os resultados. Além disso, o programa deverá informar se há mais números positivos, mais números negativos ou se há a mesma quantidade.

````python
contador_positivo = 0
contador_negativo = 0

numero = 0 # int(input("Digite um número inteiro (0 para parar): "))

while numero != 0:
    if numero > 0:
        contador_positivo += 1
    elif numero < 0:
        contador_negativo += 1


    numero = int(input("Digite outro número inteiro (0 para parar): "))


print("RESULTADO FINAL:")
print(f"Total de números positivos: {contador_positivo}")
print(f"Total de números negativos: {contador_negativo}")


if contador_positivo > contador_negativo:
    print("Foram digitados mais números positivos.")
elif contador_negativo > contador_positivo:
    print("Foram digitados mais números negativos.")
else:
    print("Foram digitados a mesma quantidade de números positivos e negativos.")
````

3) 	Faça um programa que converta a temperatura de graus Celsius para graus Farenheit ou vice-versa. O programa deverá perguntar se a temperatura de entrada está em Celsius ou Farenheit e mostrar a temperatura digitada convertida para a outra unidade. Fórmulas úteis:  $$C=\frac{5(F-32)}{9}$$,   $$F=\frac{9C}{5}+32$$.

````python
formato_temperatura = "Celsius" # input("A temperatura está em Celsius ou Fahrenheit? ").lower()

if formato_temperatura == "fahrenheit":
    fahrenheit = 32 # float(input("Qual a temperatura em Fahrenheit? "))
    celsius = (fahrenheit - 32) * 5 / 9
    print(f"A temperatura convertida é {celsius:.2f}° Celsius.")
elif formato_temperatura == "Celsius":
    celsius =  21 # float(input("Qual a temperatura em Celsius? "))
    fahrenheit = (celsius * 9 / 5) + 32
    print(f"A temperatura convertida é {fahrenheit:.2f}° Fahrenheit.")
else:
    print("Formato inválido! Digite 'Celsius' ou 'Fahrenheit'.")
````
