```python
numero_1 = int(input("digite o primeiro numero: "))
numero_2 = int(input("digite o segundo numero: "))

if(numero_1 > numero_2):
  print("o primeiro numero é maior")
elif(numero_1 < numero_2):
  print("o segundo numero é maior")
else:
  print("os numeros sao iguais")
```

```python
sexo = input("Digite seu sexo: ").lower()
idade = int(input("Digite sua idade: "))
contribuicao = int(input("Digite quantos anos de contribuição você tem: "))

if sexo == "masculino":
    if idade >= 66 or contribuicao >= 36:
        print("Você está apto a se aposentar.")
    else:
        print("Você não está apto a se aposentar.")
elif sexo == "feminino":
    if idade >= 62 or contribuicao >= 33:
        print("Você está apta a se aposentar.")
    else:
        print("Você não está apta a se aposentar.")
else:
    print("Sexo não reconhecido. Por favor, insira 'masculino' ou 'feminino'.")
```

```python
classe = input("Digite sua classe (Guerreiro ou Ladino): ").capitalize()
nivel = int(input("Digite seu nível: "))
item_especial = input("Você possui um item especial? (sim ou não): ").lower() == "sim"

titulo = "sem título"

if item_especial:
    if classe == "Guerreiro":
        if nivel >= 30:
            titulo = "Guerreiro maior"
        else:
            titulo = "Guerreiro menor"
    elif classe == "Ladino":
        if nivel >= 30:
            titulo = "Ladino maior"
        else:
            titulo = "Ladino menor"

print(f"Você é um {titulo}.")
```
