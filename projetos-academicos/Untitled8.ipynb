```python
def palavras_vogais (palavras):
  contador = 0
  for letra in palavras:
    if letra == "a":
      contador +=1
    elif letra == "e":
      contador +=1
    elif letra == "i":
      contador +=1
    elif letra == "o":
      contador +=1
    elif letra == "u":
      contador +=1
  print(f"A palavra tem {contador} vogais")

entrada = input("digite uma palavra: ")
palavras_vogais(entrada)
```

```python
def sominha(numero1, numero2):
  if(numero1 >= 0 and numero2 >= 0):
    print(numero1 + numero2)
  elif(numero1 < 0 and numero2 < 0):
    print(numero1 + numero2)
  else:
    print("os numeros tem sinais diferentes")

aaa1 = int(input("digite um numero: "))
aaa2 = int(input("digite outro numero: "))
sominha(aaa1, aaa2)
```

```python
numero1 = int(input("digite um número: "))
numero2 = int(input("digite outro numero:"))
sinal = input("qual operação voce quer fazer? ").strip().lower()
if sinal == "Soma":
  print( numero1 + numero2)
elif sinal == "Multiplicação":
  print(numero1 * numero2)
elif sinal == "Divisão":
  print(numero1 / numero2)
elif sinal == "Subtração":
  print(numero1 - numero2)
else:
  print("essa operacao nao existe tente novamente")
```
